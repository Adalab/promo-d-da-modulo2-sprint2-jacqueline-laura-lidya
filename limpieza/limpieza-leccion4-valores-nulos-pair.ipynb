2. Es el momento de eliminar los nulos:
- Reemplazad los valores nulos de la columna age por la media de la edad, redondeada a dos decimales.

```python
df_bank["age"].fillna(df_bank["age"].mean().round(2), inplace = True)
```

```python
df_bank.stb.missing()
```

- Reemplazad los valores nulos de la columna duration por la mediana de la edad, redondeada a dos decimales.

```python
df_bank["duration"].fillna(df_bank["duration"].median().round(2), inplace = True)
```

```python
df_bank.stb.missing()
```

- En relaci贸n a las columnas de education, default, housing y loan al tratarse de columnas de tipo categ贸rica, reemplazad los valores nulos por una nueva categ贸ria que se llame unknown.

```python
lista_columnas_unknown = ["education", "default", "housing", "loan"]

for i in lista_columnas_unknown:
    df_bank[i].fillna("Unknown", inplace = True)
```

```python
df_bank.stb.missing()
```

- Reemplazad los valores nulos de las columna job y marital por el valor m谩s frecuente (la moda).

```python
lista_moda = ["job", "marital"]

for i in lista_moda:
    df_bank[i].fillna(df_bank[i].mode()[0], inplace = True)
```
